```python
from dash import Dash, html, dcc, Input, Output, State
import dash_bootstrap_components as dbc
import pandas as pd

app = Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

# --- FUNÇÃO AUXILIAR PARA DEFINIR AS CORES ---
# Esta função centraliza a lógica de cores, tornando o código mais limpo.
def get_color_from_confidence(confidence_level):
    """Retorna uma cor do Bootstrap baseada no nível de confiança."""
    confidence_lower = confidence_level.lower()
    if 'alta' in confidence_lower:  # Pega "alta" e "muito alta"
        return "success"  # VERDE
    elif 'média' in confidence_lower:
        return "warning"  # AMARELO
    elif 'baixa' in confidence_lower:
        return "danger"   # VERMELHO
    return "secondary"  # Cor padrão para casos inesperados

app.layout = dbc.Container([
    dbc.Row([
        dbc.Col([
            html.Div([
                html.H1("🦟 Sistema de diagnóstico de doenças transmitidas por artrópodes", 
                       className="text-center text-white mb-0"),
                html.P("IA Híbrida: Sistema Especialista + Processamento de Linguagem Natural", 
                      className="text-center text-white-50 mb-0")
            ], style={
                'background': 'linear-gradient(135deg, #17ead9 0%, #6078ea 100%)', # Novo gradiente de cor
                'padding': '30px',
                'borderRadius': '10px',
                'marginBottom': '20px'
            })
        ], width=12)
    ]),
    
    dbc.Row([
        dbc.Col([
            dbc.Card([
                dbc.CardHeader(html.H4("📝 Descrição do Caso", className="mb-0")),
                dbc.CardBody([
                    dbc.Textarea(
                        id="input-texto",
                        placeholder="Descreva os sintomas do paciente em texto livre...\n\nExemplo: 'Paciente com febre alta há 3 dias, dor de cabeça intensa, dor atrás dos olhos e manchas vermelhas na pele. Visto mosquito em casa.'",
                        style={'height': '200px', 'fontSize': '14px'},
                        className="mb-3"
                    ),
                    html.Div([
                        dbc.Button("🔍 Analisar e Diagnosticar", id="btn-diagnosticar", 
                                 color="primary", size="lg", className="me-2"),
                        dbc.Button("🗑️ Limpar", id="btn-limpar", 
                                 color="secondary", size="lg", outline=True)
                    ])
                ])
            ], className="mb-4")
        ], width=12)
    ]),
    
    dbc.Spinner([
        html.Div(id="output-diagnostico")
    ], color="primary", type="border")
    
], fluid=True)

@app.callback(
    Output("output-diagnostico", "children"),
    Input("btn-diagnosticar", "n_clicks"),
    State("input-texto", "value"),
    prevent_initial_call=True
)
def realizar_diagnostico(n_clicks, texto):
    if not texto or texto.strip() == "":
        return dbc.Alert("⚠️ Por favor, descreva os sintomas do paciente.", color="warning")
    
    try:
        resultado = motor.diagnosticar(texto)
        output = []
        
        # 1. FATOS EXTRAÍDOS
        fatos = resultado['fatos_extraidos']
        output.append(
            dbc.Card([
                dbc.CardHeader(html.H5("🔍 Fatos Extraídos pelo NLP")),
                dbc.CardBody([
                    html.P([html.Strong("Sintomas identificados: "), 
                           f"{len(fatos['sintomas'])} sintomas"]),
                    html.Ul([html.Li(s) for s in fatos['sintomas']]) if fatos['sintomas'] else html.P("Nenhum sintoma claro identificado."),
                    html.Hr(),
                    html.P([html.Strong("Vetor: "), fatos['vetor'] or "Não identificado"]),
                    html.P([html.Strong("Viagem: "), "Sim" if fatos['menciona_viagem'] else "Não"])
                ])
            ], className="mb-3", color="light")
        )
        
        # 2. DIAGNÓSTICOS
        diagnosticos = resultado['diagnosticos']
        if diagnosticos:
            output.append(
                dbc.Card([
                    dbc.CardHeader(html.H5("🎯 Diagnósticos Possíveis")),
                    dbc.CardBody([
                        html.Div([
                            dbc.Alert([
                                html.H4(f"{i+1}. {diag['doenca']}", className="alert-heading"),
                                html.Hr(),
                                html.Strong("Probabilidade:"),
                                dbc.Progress(
                                    value=diag['probabilidade']*100,
                                    label=f"{diag['probabilidade']*100:.1f}%",
                                    className="mb-2",
                                    # **MUDANÇA APLICADA AQUI**
                                    color=get_color_from_confidence(diag['confianca'])
                                ),
                                html.P([
                                    html.Strong("Confiança: "), 
                                    dbc.Badge(diag['confianca'].upper(), 
                                             # **MUDANÇA APLICADA AQUI**
                                             color=get_color_from_confidence(diag['confianca']),
                                             className="me-2"),
                                    dbc.Badge(f"Método: {diag['metodo']}", color="info")
                                ]),
                                # Demais informações sobre a doença
                                html.Ul([
                                    html.Li(f"Vetor: {BASE_CONHECIMENTO.get(diag['doenca'].lower().replace(' ', '_'), {}).get('vetor', 'N/A')}"),
                                    html.Li(f"Gravidade: {BASE_CONHECIMENTO.get(diag['doenca'].lower().replace(' ', '_'), {}).get('gravidade', 'N/A')}")
                                ], style={'fontSize': '0.9em'})
                            ], 
                            # **MUDANÇA APLICADA AQUI** (cor do card principal)
                            color=get_color_from_confidence(diag['confianca']) if i == 0 else "light")
                            for i, diag in enumerate(diagnosticos)
                        ])
                    ])
                ], className="mb-3")
            )
        else:
            output.append(dbc.Alert("⚠️ Não foi possível determinar um diagnóstico com os sintomas informados.", color="warning"))
        
        # 3. RECOMENDAÇÕES E AVISO
        output.append(dbc.Card([
            dbc.CardHeader(html.H5("💡 Recomendações")),
            dbc.CardBody([html.Ul([html.Li(rec) for rec in resultado['recomendacoes']])])
        ], className="mb-3", color="primary", outline=True))
        
        output.append(dbc.Alert([
            html.H5("⚠️ AVISO IMPORTANTE", className="alert-heading"),
            "Este sistema é uma ferramenta de apoio e NÃO substitui a consulta médica. Sempre procure um profissional para o diagnóstico definitivo."
        ], color="warning"))
        
        return output
        
    except Exception as e:
        return dbc.Alert(f"❌ Erro ao processar: {str(e)}", color="danger")

@app.callback(
    Output("input-texto", "value"),
    Input("btn-limpar", "n_clicks"),
    prevent_initial_call=True
)
def limpar_texto(n_clicks):
    return ""
```
